```python
# What is the churn distribution among the customers?
Churn_distribution = data["Churn"].value_counts()
Churn_distribution
```

```python
# How much time do churned users and non churned users spend on the network during the day, evening and night on average?
Average_time_spent = ["Total day minutes", "Total eve minutes", "Total night minutes"]

data.groupby(["Churn"])[Average_time_spent].mean()
```

```python
# What is the average total charge (Day, evening, night, international) for customers who churned versus those who did not?
Average_total_charge = ["Total day charge", "Total eve charge", "Total intl charge", "Total night charge"]

data.groupby(["Churn"])[Average_total_charge].mean()
```

```python
# Are there significant differences in churn rates across different states or area codes?
Churn_rate_areacode = ["Area code"]

data.groupby(Churn_rate_areacode)["Churn"].mean()
```

```python
# Are there significant differences in churn rates across different states or area codes?
Churn_rate_state = ["State"]

data.groupby(Churn_rate_state)["Churn"].mean()
```

```python
# Do customers with an international plan have different usage patterns compared to those without?
Usage_pattern = ["Total day minutes", "Total eve minutes", "Total night minutes"]

data.groupby(["International plan"])[Usage_pattern].mean()
```

```python
# Who engages customer care the most among the customers?
data.groupby(["International plan", "Churn"])["Customer service calls"].mean()
```

```python
# How many customers are on international plan and how many are not
International_plan_customers = data["International plan"].value_counts()
International_plan_customers
```

```python
# How many customers have voicemail plans and how many do not have?
voicemail_plan_customers = data["Voice mail plan"].value_counts()
voicemail_plan_customers
```
